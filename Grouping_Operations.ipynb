**Groupung Operations**

In most machine learning algorithms, every instance is represented by a row in the training dataset, where every column show a different feature of the instance. This kind of data called “Tidy”.

Datasets such as transactions rarely fit the definition of tidy data above, because of the multiple rows of an instance. In such a case, we group the data by the instances and then every instance is represented by only one row.

The key point of group by operations is to decide the aggregation functions of the features.

![alt text](https://miro.medium.com/max/1400/1*VWBbZRkTrHJQrQfWlPQWUg.png)

We can see that the data is group by user so it reduces the number of rows and the data become more tidy and understandable.


There are two methods for grouping :



1.   Group by method
2.   Pivot table




````
import pandas as pd
import numpy as np
````

````
df=pd.read_csv('/content/student-mat.csv')
````

````
df.school.unique()
````

**Group by**

````
g=df.groupby('sex') 
````

````
# 1 row of each group
g.first()
````

````
#All row of specefic sub group
g.get_group('F') 
````

````
##Grouping with multiple columns
gkk = df.groupby(['sex', 'health']) 
````

````
gkk.first()
````

````
gkk.get_group(('M',1))
````

**Using Pivot Table** 

General code
```
data.pivot_table(index='column_to_group', columns='column_to_encode', values='aggregation_column', aggfunc=np.sum, fill_value = 0)
```

````
## Basic pivot table
df.pivot_table(index=['health','sex'])
````

````
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df['paid'] = label_encoder.fit_transform(df['paid'])
````

````
df.paid
````

````
df.pivot_table(index=['health','sex'],values=['paid'], aggfunc=np.sum)
````

From above table we can see that number of people with health =1, sex=Female who have paid fees are 18. 

````
# same table using column and index
df.pivot_table(index='health',columns='sex',values=['paid'], aggfunc=np.sum)
````
